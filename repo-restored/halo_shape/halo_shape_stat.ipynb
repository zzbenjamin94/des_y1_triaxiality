#### Resample randomly selected halos to match mass distribution of redMaPPer selected halos.
Ellipticity is correlated with mass, more massive more elliptical. When comparing ellipticity between random and redMaPPer need to tease out the confounding variable of mass distribution. Random follows halo mass function, while redMaPPer peaks at $10^{14}~M_{\odot}$. 

```python
from numpy.random import choice
'''
Resamples the randomly selected halos to match the mass distribution of redMaPPer selected halos. 
The mass is resampled to match a mass histogram between 10^13 - 10^15 M_sun and with number of bins defined
by the user. The lower and upper limit is to prevent the sparse samples to choose from beyond these limits,
which in the incident of no samples to choose from in a bin the function terminates with an error message. 

The function collected the indices of the resampled halos and outputs the shape statistics with the resampled 
halos. It is possible to resample a halo more than once. 

Input:
halos_M: masses of the halos with distribution which the resampled random halos follow, in Msun/h. 
halos_M_rand: masses of randomly sampled halos in Msun/h
num_bin: Number of logarithmically spaced bins between the lower and upper mass limits. Default to 30.

Output:
rand_resamp_ind: Array of indices of the halos_M_rand array of the resampled halos.
'''

def halos_rand_resamp(halos_M, halos_M_rand, num_bin = 30):
    m_min = 13; m_max = 15   
    n_mass_hist, lnmass_bin_edge = np.histogram(np.log10(halos_M), num_bin, range=(m_min, m_max))

    #Resampled statistics
    rand_resamp_ind = np.array([], dtype=int)
    for i in range(num_bin):
        lnmass_bin_low = lnmass_bin_edge[i]; lnmass_bin_high = lnmass_bin_edge[i+1]
        lnmass_bin_ind = np.where( (np.log10(halos_M_rand) < lnmass_bin_high) & \
                             (np.log10(halos_M_rand) >= lnmass_bin_low))[0]

        #Draw samples from randomly selected halos to match redMapper.
        #The two distributions may be off at large mass, so introduce arbitary cutoff at 10**13 and 10**15 Msun
        #Should not affect distribution too much as >99% are in the 13-15 range

        #Resample the distribution in the mass range to match number of halos in redMapper
        assert len(lnmass_bin_ind) >= 1, "No halos in random in  mass range ({},{})".format(lnmass_bin_low,lnmass_bin_high)
        lnmass_bin_resamp_ind = choice(lnmass_bin_ind, size=n_mass_hist[i])
        rand_resamp_ind = np.concatenate((rand_resamp_ind, lnmass_bin_resamp_ind))

    return rand_resamp_ind


```

```python
#Resmpling the random halo statistics
rand_resamp_ind = halos_rand_resamp(halos_M, halos_M_rand)
halos_M_rand = halos_M_rand[rand_resamp_ind]
q_rand = q_rand[rand_resamp_ind]
s_rand = s_rand[rand_resamp_ind]
halos_ID_rand = halos_ID_rand[rand_resamp_ind]
cos_i_rand = cos_i_rand[rand_resamp_ind]
halos_RA_rand = halos_RA_rand[rand_resamp_ind]
halos_DEC_rand = halos_DEC_rand[rand_resamp_ind]

#coordinates to input to kmeans
halos_coord_rand = np.zeros( (len(halos_RA_rand),2))
halos_coord_rand[:,0] = halos_RA_rand; halos_coord_rand[:,1] = halos_DEC_rand
```
